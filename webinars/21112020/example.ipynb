## Задача 1

Медведь Винни-Пух арендовал $3$ помещения для хранения мёда. 
Каждое помещение заперто на замок, и у медведя есть по $2$ ключа от 
каждого из замков. Ключи выглядят одинаково, поэтому Винни-Пух 
каждый раз перебирает их для того, чтобы открыть дверь. 
Один и тот же ключ медведь несколько раз не пробует. Если медведь 
не открывает дверь за $3$ попытки, то он расстраивается и впадает в спячку. 
Найди тевероятность того, что сегодня Винни-Пух доберётся до мёда.

### Решение 1

С вероятностью $P_1 = \dfrac{4}{6}$ Винни-Пух не открывает дверь. Учитвая, что Винни-Пух не кладет обратно неудачный ключ, то

С вероятность $P_2 = P_1 \times \dfrac{3}{5} =  \dfrac{4}{6} \times \dfrac{3}{5}$ Винни-Пух не открывает вторую дверь, аналогично получаем вероятность неудачи трех попыток - ответ:

$
P_3 = P_2 \times \dfrac{2}{4} = \dfrac{4}{6} \times \dfrac{3}{5} \times \dfrac{2}{4} = \dfrac{1}{5} = 0.2
$

### Решение 2

Всего вариантов перестановок $A_6^6 = 6!$

Неудачи состоят из:

- $3$ неудачи из $4$: $A_4^3 = \dfrac{4!}{(4-3)!} = 4!$

- $3$ оставшихся случая из $6$: $A_3^3=3!$

Тогда ответ:

$
P_3 = \dfrac{A_4^3A_3^3}{A_6^6} = \dfrac{4!3!}{6!} = \dfrac{1}{5} = 0.2
$

## Задача 2

Какая будет функция распределения величины “количество опробованных ключей допервого успеха” из предыдущей задачи?

### Решение

С вероятностью $P_1 = \dfrac{2}{6} = \dfrac{1}{3}$ - дверь открывается с первой попытки

С вероятностью $P_2 = \bigg(1-P_1\bigg)\dfrac{2}{5} + P_1 = \dfrac{2}{3}\dfrac{2}{5} + \dfrac{1}{3} = \dfrac{3}{5}$ - дверь открывается со второй попытки

С вероятностью $P_3 = \bigg(1-P_2\bigg)\dfrac{2}{4} + P_2 = \dfrac{2}{5}\dfrac{2}{4} + \dfrac{3}{5} = \dfrac{4}{5}$ - дверь открывается с третьей попытки

С вероятностью $P_4 = \bigg(1-P_3\bigg)\dfrac{2}{3} + P_3 = \dfrac{1}{5}\dfrac{2}{3} + \dfrac{4}{5} = \dfrac{14}{15}$ - дверь открывается с четвертой попытки

С вероятностью $P_5 = \bigg(1-P_4\bigg)\dfrac{2}{2} + P_4 = (1-P_4) + P_4 = 1$ - дверь открывается с пятой попытки

## Задача 3

Теперь предположим, что медведь не такой сообразительный и он может пробоватьодин и тот же ключ несколько раз. Он берёт случайный ключ, пробует его, если он неподходит, то он кладёт его обратно и повторяет процедуру. Найдите матожиданиеколичества попыток.

### Решение

Пусть $q = 1 - p$, тогда, исходя из условия, вероятность угадать ключ с $n^{\textrm{й}}$ попытки равна 

$P_n = q^{n-1}p$

Сумма всех вероятностей:

$P = \sum\limits_{i=1}^n P_i = \sum\limits_{i=1}^n q^{i-1}p = \dfrac{p}{1-q} = 1$

Продифференцируем ее по $q$:

С одной стороны:

$\dfrac{\partial P}{\partial q} = \sum\limits_{i=1}^n iq^{i-1}p$ 

Это математическое ожидание.

С другой стороны: 

$\dfrac{\partial P}{\partial q} = \dfrac{p}{(1-q)^2}$

Отсюда ответ: 

$E(n) = \dfrac{p}{(1-q)^2} = \dfrac{1}{p} = 3$

## Задача 4

Винни-Пух добрался до мёда и теперь торгует им на бирже. Цена мёда каждый день растёт на $1$ рубль с вероятностью $50%$, падает на $1$ рубль с вероятностью $30%$ и не меняется с вероятностью $20%$. Найдите вероятность того, что цена мёда подниметсяровно на $2$ рубля за $5$ дней.

### Решение

Всего имеется два случая, которые нас удовлетворяют:

- 2 раза цена растет и 3 раза не меняется:

$P_1 = C_2^5\times0.5^2\times C_3^3\times0.2^3$

- 3 раза цена растет, 1 раз уменьшается и 1 раз падает

$P_2 = C_3^5\times0.5^3\times2\times0.3\times0.2$

Тогда ответ:

$P = P_1 + P_2 = 0.17$

## Задача 5

Начальная цена мёда равна $5$ рублям. Каждый день она поднимается на рубль с
вероятностью $0.7$ и не меняется с вероятностью $0.3$ . Найдите вероятность того, что
цена вырастет ровно вдвое за $10$ дней.

### Решение

Тут надо просто расставить дни и умножить на вероятности.

Ответ:

$P = C_5^{10}\times0.7^5\times0.3^5 = 0.1029193452$
