# Задание 5.2

Среди тех, кому показали рекламный баннер, $20$% оказались мужчинами. Мужчины переходят по ссылке баннера с вероятностью $5$%, а женщины — с вероятностью $2.5$%

1) С какой вероятностью пользователь, которому показали рекламу, окажется женщиной, кликнувшей по ссылке?

$P = (1-P(\text{М}))P(\text{Ж}\cap1) = 0.8 * 0.025 = 0.02$

2) Выберите, каких пользователей больше:

$P(\text{Ж}\cap1) = (1-P(\text{М}))P(\text{Ж}|1) = 0.8 * 0.025 = 0.02$

$P(\text{М}\cap1) = P(\text{М})P(\text{М}|1) = 0.2 * 0.05 = 0.01$

$\Rightarrow P(\text{Ж}\cap1) > P(\text{М}\cap1)$

Ответ: женщин, кликнувших по ссылке.
