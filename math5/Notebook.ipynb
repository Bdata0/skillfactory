# Blank Advanced Problem (7)

$2$% клиентов испытали wow-effect от вашего сервиса и рассказывают о вас буквально всем. Из них $70$% приведут новых клиентов. Еще $20$% клиентов периодически оставляют вам положительные отзывы в соцсетях и на сайте. Из них $10$% приведут новых клиентов. Что думают оставшиеся $78$% клиентов, вы не знаете. Из них $4$% приведут новых клиентов.

1) Выберите верное утверждение:

Пусть: 
- $A$ - событие, что клиент привел новых клиентов
- $B$ - событие, что клиент испытал wow-effect
- $C$ - событие, что клиент периодически оставляет нам положительные отзывы
- $D$ - событие, что мы не знаем клиента

Тогда

$P(\text{«клиент приведет нового клиента»}|\text{«клиент испытал wow-effect»})  = P(A|B) = 0.7$ - верное утверждение

2) Если вы хотите найти вероятность, что случайный клиент вашего сервиса приведет нового клиента, что вы возьмёте за разбиение?

Типы клиентов верно

3) Найдите вероятность, что случайный клиент вашего сервиса приведет нового клиента

$P(A) = \sum\limits_{i=\{B,C,D\}}{P(A|i)P(i)} = 0.02\times0.7+0.2\times0.1+0.78\times0.04 = 0.0652$

4) Какой из типов клиентов дает наибольший вклад в приведение новых клиентов

Клиенты-тихони, про которых ничего не известно

5) Кто-то из ваших клиентов привел нового клиента. Какая вероятность, что это сделал клиент тихоня?

$P(D|A) = \dfrac{P(D)P(A|D)}{P(A)}$

# Задача 8.2

$2$% клиентов испытали wow-effect от вашего сервиса и рассказывают о вас буквально всем. Из них $70$% приведут новых клиентов. Еще $20$% клиентов периодически оставляют вам положительные отзывы в соцсетях и на сайте. Из них $10$% приведут новых клиентов. Что думают оставшиеся $78$% клиентов, вы не знаете. Из них $4$% приведут новых клиентов.

Пусть: 
- $A$ - событие, что клиент привел новых клиентов
- $B$ - событие, что клиент испытал wow-effect
- $C$ - событие, что клиент периодически оставляет нам положительные отзывы
- $D$ - событие, что мы не знаем клиента

Тогда

$P(A) = \sum\limits_{i=\{B,C,D\}}{P(A|i)P(i)} = 0.02\times0.7+0.2\times0.1+0.78\times0.04 = 0.0652$

1) Вычислите апостериорную вероятность принадлежности клиента к классу «wow-effect», если стало известно, что он привел нового клиента. Ответ округлите до сотых.

$P(B|A) = \dfrac{P(B)P(A|B)}{P(A)} \approx 0.22$

2) Вычислите апостериорную вероятность принадлежности клиента к классу «тихоня», если стало известно, что он привел нового клиента. Ответ округлите до сотых.

$P(D|A) = \dfrac{P(D)P(A|D)}{P(A)} \approx 0.48$

3) Вычислите апостериорную вероятность принадлежности клиента к классу «тихоня», если известно, что он НЕ привел нового клиента. Ответ округлите до сотых.

$P(D|\overline A) = \dfrac{P(D)P(\overline A|D)}{P(\overline A)} = \dfrac{P(D)(1-P(A|D))}{1 - P(A)} \approx 0.80$

# Задача 8.3

Редкое заболевание встречается у $1$% людей. Тест ошибочно показывает наличие болезни у здоровых людей с вероятностью $2$% и ошибочно показывает отсутствие болезни у больных с вероятностью $3$%. Найдите вероятность, что человек на самом деле болен, если тест отрицателен на болезнь.

Пусть

- $A$ - человек болен
- $B$ - тест показал наличие болезни

Тогда

$P(\overline B) = P(\overline A)(1-P(B|A)) + (1-P(\overline A))P(\overline B|A) = 0.99\times0.98+0.01\times0.03 = 0.9705$

Получим

$P(A|\overline B) = 1-\bigg(\dfrac{P(\overline A)P(\overline B|\overline A)}{P(\overline B)}\bigg) \approx 0.0003$

Ответ: 0.0003
